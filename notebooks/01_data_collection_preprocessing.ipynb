```python
import pandas as pd
import numpy as np

from google_play_scraper import reviews, Sort
```

```python
cbe_app_id = "com.combanketh.mobilebanking"
boa_app_id = "com.boa.boaMobileBanking"
dashen_app_id = "com.dashen.dashensuperapp"
```

```python
def scrape_reviews(app_id, bank_name, count=800):
    
    result, continuation_token = reviews(
        app_id,
        lang='en',
        country='et',
        sort=Sort.NEWEST,
        count=count
    )
    
    df = pd.DataFrame(result)
    
    df['bank'] = bank_name
    
    return df
```

```python
cbe_reviews = scrape_reviews(cbe_app_id, "CBE")
boa_reviews = scrape_reviews(boa_app_id, "BOA")
dashen_reviews = scrape_reviews(dashen_app_id, "Dashen")
```

# Commercial Bank of Ethiopia Reviews

```python
cbe_reviews.head()
```

```python
cbe_reviews = cbe_reviews[
    [
        'reviewId',
        'userName',
        'content',
        'score',
        'thumbsUpCount',
        'at',
        'bank'
    ]
]
```

```python
cbe_reviews.columns = [
    'review_id',
    'user_name',
    'review',
    'rating',
    'thumbs_up',
    'review_date',
    'bank'
]
```

```python
cbe_reviews.isnull().sum()
```

```python
cbe_reviews.dropna(subset=['review'], inplace=True)
```

```python
cbe_reviews.dropna(subset=['review'], inplace=True)
```

```python
cbe_reviews.drop_duplicates(subset=['review'], inplace=True)
```

```python
import re

def clean_text(text):
    
    text = text.lower()
    
    text = re.sub(r"http\\S+", "", text)
    
    text = re.sub(r"[^a-zA-Z\\s]", "", text)
    
    text = re.sub(r"\\s+", " ", text).strip()
    
    return text
```

```python
cbe_reviews['clean_review'] = cbe_reviews['review'].apply(clean_text)
```

```python
cbe_reviews.to_csv(
    "../data/processed/cbe_reviews_clean.csv",
    index=False
)
```
